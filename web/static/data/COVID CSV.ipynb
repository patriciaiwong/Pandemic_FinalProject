```python
covid = "data.csv"
covid_df = pd.read_csv(covid)
covid_df.head()
```

```python
del covid_df['city']
covid_df
```

```python
covid_df.count()
```

```python
columns = ["county", "state", "country", "cases", "deaths", "lat", "long"]
covid_usa = covid_df.loc[covid_df["country"] == "USA", columns]
covid_usa
```

```python
# usa_index = covid_usa.reset_index(drop=True)
cleanup = covid_usa.dropna(subset=['state'])
```

```python
cleanup = cleanup.dropna(subset=['lat'])
```

```python
cleanup = cleanup.dropna(subset=['long'])
```

```python
cleanup = cleanup[cleanup['cases'] != 0]
```
